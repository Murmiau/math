```python
1. Когда используется критерий Стьюдента, а когда Z –критерий?
```

```python
T-тесты используются для сравнения средних значений двух групп, когда стандартное отклонение совокупности неизвестно, 
а Z-тесты используются, когда известно стандартное отклонение совокупности и размер выборки большой.
T-тесты основаны на t-распределении, которое используется для меньших размеров выборки и неизвестных стандартных отклонений 
населения, тогда как Z-тесты используют стандартное нормальное распределение.
На практике t-тесты более распространены из-за редкости известных стандартных отклонений населения. В то же время Z-тесты, как 
правило, предназначены для ситуаций с большими размерами выборки и известными параметрами населения.
```

```python
2. Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 
17 мм. Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр 
оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.
```

```python
Н0: M = M0 = 17
H1: M > M0
т.к. известны и среднее, и дисперсия генеральной совокупности, то для проверки гипотезы воспользуемся z-критерием.
```

```python
z=(17.5-17)/0.2
print(f'Наблюдаемое значение z-критерия: {z: .4f}')
zt=stats.norm.ppf(0.95)
print(f'Табличное значение z-критерия: {zt: .4f}')
if(z>zt):
    print("=> отвергаем нулевую гипотезу на уровне значимости alpha=0.05 => заявление изготовителя не подтверждается")
else:
    print("=> нулевую гипотезу не отвергаем на уровне значимости alpha=0.05 => подтверждается заявление изготовителя")
```

```python
3. Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья составляет 200 г.
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.
Известно, что их веса распределены нормально.

Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%? (Провести двусторонний тест.)
```

```python
Проводим проверку следующих гипотез:
Н0: M =200
H1: M не равно 200

т.к. не известны ни среднее, ни дисперсия генеральной совокупности, то для проверки гипотезы воспользуемся t-критерием.
```

```python
weight = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
n = 10

mean = weight.mean()
std = weight.std(ddof=1)

mean, std
```

```python
t = (mean-200)/(std/np.sqrt(n))
t
```

```python
4. Есть ли статистически значимые различия в росте дочерей?
Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169
Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160
```

```python
mothers=np.array([172,177,158,170,178,175,164,160,169,165])
daughters=np.array([173,175,162,174,175,168,155,170,160,163])

def equalmean(x,y):
    dif=mothers-daughters
    n=len(dif)
    mean_dif=dif.mean()
    var_dif=dif.var()*n/(n-1)
    t=mean_dif/math.sqrt(var_dif)*math.sqrt(n)
    p = 1 - math.fabs(stats.t.cdf(t,n-1) - stats.t.cdf(-t,n-1))
    return t,p
    
t,p=equalmean(mothers,daughters)
print("t-статистика критерия = ",t)
print("p-уровень = ",p)
alpha=0.05
if p>alpha:
    print("Статистически значимых различий в росте матерей и дочерей нет.")
else:
    print("Статистически значимые различия в росте матерей и дочерей есть.")
```
