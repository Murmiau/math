```python
import numpy as np
import pandas as pd
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import pandas as pd
```

```python
Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и 
штангистов. Даны значения роста в трех группах случайно выбранных спортсменов: 
Футболисты: 173, 175, 180, 178, 177, 185, 183, 182. 
Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180. 
Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.
```

```python
football = np.array([173, 175, 180, 178, 177, 185, 183, 182])
hockey = np.array([177, 179, 180, 188, 177, 172, 171, 184, 180])
weightlifter = np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170])
```

```python
stats.kruskal(football, hockey, weightlifter) 
```

```python
df = pd.DataFrame({"height": [173, 175, 180, 178, 177, 185, 183, 182, 
                             177, 179, 180, 188, 177, 172, 171, 184, 180,
                             172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170],
                   "group": np.repeat(["football", "hockey", "weightlifter"], repeats = [8, 9, 11])})
df
```

```python
tukey = pairwise_tukeyhsd(df["height"], df["group"], alpha = 0.05)
print(tukey)
```

```python
Вывод => между группой футболистов и хоккеистов нет различий в среднем росте, в свою очередь между группами футболистов и 
штангистов, и между группами хоккеистов и штангистов есть различия в среднем росте.
```
